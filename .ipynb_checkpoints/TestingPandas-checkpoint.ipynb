```python

pd.set_option('display.max_rows', 200)
filtered_df = df[df['name'] == 'United States']
columns_to_view=['date', 'local_price', 'name']
final_view=filtered_df[columns_to_view]
print(final_view)
```

```python

filtered_df = df[df['name'] == 'United States']
columns_to_view=['date', 'local_price', 'name']
final_view=filtered_df[columns_to_view]

price_by_date=filtered_df.groupby('date')['local_price'].sum().reset_index()
plt.figure(figsize=(8,5))
plt.bar(price_by_date['date'], price_by_date['local_price'])
plt.xlabel('DATE')
plt.ylabel('Todal Price in $')
plt.title('Total Price for United States by Date')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
#plt.savefig('total_price_by_date_bar.png')
plt.show()
```
